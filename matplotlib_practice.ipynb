```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Alternative to line above
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Practice with bar charts.
plt.bar(x_axis, y_axis)
```

```python
# Practice with attributes
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Labels for x and y
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Title
plt.title("PyBer Fare by Month")
# Legend
plt.legend()
```

```python
# Horizontal Bar Chart
plt.barh(x_axis, y_axis)
```

```python
# Put months in correct ascending order
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Vertical bar charts using object-oriented interface method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Horizontal bar using object-oriented
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Scatter plot practice
plt.plot(x_axis, y_axis, 'o')
# OR --> plt.scatter(x_axis, y_axis)
```

```python
# Bubble chart practice
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Using Pythons list comprehension technique, optimized to spot a predictable
# pattern during looping instead of many 'for' and 'while' loops
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Object-oriented scatter plots
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Object-oriented bubble chart 
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Pie charts using matlab
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Object-oriented pie chart
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```
